### Load Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split 

from sklearn.datasets import load_iris
from sklearn.datasets import load_breast_cancer
```

### Iris Dataset

```python
iris_ds = load_iris()
```

```python
trials = 50

# store accuracies for each n_neighbors value
avg_train_acc = []
avg_test_acc = []

for i in range(trials):
    X_train, X_test, y_train, y_test = train_test_split (
        iris_ds.data, iris_ds.target, test_size=0.25, random_state=i)
    
    # store accuracies for each trials
    train_acc = []
    test_acc = []

    neighbors_settings = range(1, 41)

    for n_neighbors in neighbors_settings:
        clf = KNeighborsClassifier(n_neighbors=n_neighbors)
        clf.fit(X_train, y_train)
        
        # train set accuracy
        train_acc.append(clf.score(X_train, y_train))
        
        # test set accuracy
        test_acc.append(clf.score(X_test, y_test))

    # average accuracies for trial
    avg_train_acc.append(train_acc)
    avg_test_acc.append(test_acc)

# average accuracies across all trials
avg_train_acc = np.mean(avg_train_acc, axis=0)
avg_test_acc = np.mean(avg_test_acc, axis=0)

# plotting
plt.figure(figsize=(9,6))
plt.plot(avg_train_acc, label="Train accuracy")
plt.plot(avg_test_acc, label="Test accuracy")
plt.xticks(ticks=range(len(neighbors_settings)), labels=list(neighbors_settings))
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.title("Mean Plot of Accuracy vs n_neighbors with 50 trials")
plt.legend()
plt.show()
```

Upon observation, it appears that the optimal number of neighbors is 11, 13 and 14. This choice is based on the balanced performance between train and test accuracies observed at this point in the plot. I'm gonna try to validate my observation. 

```python
# Identify optimal number of neighbors
optimal_n = neighbors_settings[np.argmax(avg_test_acc)]  # Get the value of n_neighbors corresponding to the maximum test accuracy
best_test_accuracy = np.max(avg_test_acc)
best_train_accuracy = avg_train_acc[optimal_n - 1]  # Get train accuracy at the optimal n_neighbors

# Plotting
plt.figure(figsize=(9,6))
plt.plot(avg_train_acc, label="Train accuracy")
plt.plot(avg_test_acc, label="Test accuracy")
plt.axvline(x=optimal_n - 1, color='red', linestyle='--', label=f'Optimal n_neighbors ({optimal_n})')
plt.xticks(ticks=range(len(neighbors_settings)), labels=list(neighbors_settings)) 
plt.xlabel("n_neighbors")
plt.ylabel("Accuracy")
plt.title("Mean Plot of Accuracy vs n_neighbors with 50 trials")
plt.legend()
plt.show()

print(f'Optimal n_neighbors = {optimal_n}')
print(f'Optimal Neighbor Train Accuracy = {best_train_accuracy:.4f}')
print(f'Optimal Neighbor Test Accuracy = {best_test_accuracy:.4f}')
```

It appears that the best choice for the number of neighbors is 13. At this value, we achieved a high training accuracy of approximately 97.61% and a test accuracy of about 96.84%.
