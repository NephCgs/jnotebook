# Malaria Deaths Over Time ðŸ¦ŸðŸ¦ŸðŸ¦Ÿ

```python
# Philippines
df_phil = df[df['country'] == 'Philippines']
```

```python
# global average of malaria each year
df_global_total = df.groupby('year')['malaria'].sum().reset_index()

# plot the global total malaria
fig1 = px.line(df_global_total, x='year', y='malaria', title='Total Malaria Cases Over Time (Global)',
               labels={'malaria': 'Total Malaria Cases', 'year': 'Year'})

# plot malaria cases (Philippines only)
fig2 = px.line(df_phil, x='year', y='malaria', title='Malaria Cases Over Time (Philippines)',
               labels={'malaria': 'Malaria Cases', 'year': 'Year'})

# show plot
fig1.show()
fig2.show()
```

**Global**


*   **In the 1990s**, there was a noticeable increase in global malaria cases, and peaked around 2004.
*   **After the 2000s**, global cases started to decline, could be due to a boom in international efforts and campaigns by WHO, UNICEF, etc.

**Philippines**


*   The graph begins with a high number of malaria cases, then around the mid 90s we see a gradual decline.
*   Then we see a gradual increase from 2001 to 2007
*   Then a sharp decline from 2007 onwards that remained consistent throughout the years with little to no fluctuation.
*   Apparently, Australia's AusAID began funding malaria campaigns in the Philippines as part of the Roll Back Malaria (RBM) program which started around the 1990's.
*   Also based on the reports that I have read, the decline that we see on the graph from 2007 onwards aligns with the reported 74.8% reduction in malaria cases in RBM supported areas.












# Drowning Cases

```python
# Calculate global average drowning cases by year
df_global_avg = df.groupby('year')['drowning'].mean().reset_index()

# Plot global average drowning cases over time
fig1 = px.line(
    df_global_avg,
    x='year',
    y='drowning',
    title='Global Average Drowning Cases Over Time',
    labels={'drowning': 'Average Drowning Cases', 'year': 'Year'},
    line_shape='spline',
    markers=True
)
fig1.update_layout(template='plotly_dark', title_x=0.5)
fig1.show()
```

This graph shows the global average drowning cases over time from around 1990 to 2020 it indicates a steady decline in the number of drowning cases worldwide during this period.


*   In the early 1990s, global drowning cases were relatively high but started to decline gradually throughout the decade.
*   After 2000, the decline in global drowning cases became more pronounced, potentially due to increased international efforts and awareness campaigns focusing on water safety, such as those led by organizations like WHO and UNICEF.
*  From 2010 onwards, the number of drowning cases continued to decrease, though at a slower rate, suggesting that while progress has been made, additional measures may be needed to maintain the downward trend.








```python
# Plot drowning cases over time (Philippines)
fig2 = px.line(
    df_phil,
    x='year',
    y='drowning',
    title='Drowning Cases Over Time (Philippines)',
    labels={'drowning': 'Drowning Cases', 'year': 'Year'},
    line_shape='spline',  # Smooth line for better visualization
    markers=True  # Add markers to data points
)
fig2.update_layout(template='plotly_dark', title_x=0.5)
fig2.show()
```

This graph shows the number of drowning cases over time in the Philippines from around 1990 to 2020, compared to the global trend, this graph exhibits more fluctuations in the number of drowning cases.

# Tuberculosis Cases

```python
# Calculate global average tuberculosis cases by year
df_global_avg_tb = df.groupby('year')['tuberculosis'].mean().reset_index()

# Plot global average tuberculosis cases over time
fig = px.line(
    df_global_avg_tb,
    x='year',
    y='tuberculosis',
    title='Global Average Tuberculosis Cases Over Time',
    labels={'tuberculosis': 'Average Tuberculosis Cases', 'year': 'Year'},
    line_shape='spline',
    markers=True
)

fig.update_layout(template='plotly_dark', title_x=0.5)

# Show plot
fig.show()
```

Trends Over Time Look at how the line moves. An upward trend indicates an increase in average tuberculosis cases over the years, while a downward trend shows a decrease. If the line is relatively flat, it suggests that the number of cases has remained stable.

Trend Analysis- If the plot shows a steady decline in the global average tuberculosis cases over the years, it could suggest that public health measures, medical advancements, or increased awareness and diagnosis have been effective. Conversely, if the trend is upward or remains high, it could indicate that tuberculosis remains a persistent public health challenge, possibly due to factors such as drug resistance, population growth, or insufficient healthcare access in certain regions.

 - The graph shows a decline in tuberculosis cases over the given period.

 - Around the early 1990s, the number of cases was relatively high, slightly peaking in 1992 before starting a steady decline.

 - From around 1995 onward, we can see a downward trend in the number of TB cases.

 - By 2020, the average number of TB cases has significantly decreased, showing a consistent reduction over the years.

 - Overall: The graph suggests that there has been significant progress in reducing the global average of tuberculosis cases over the last few decades, which may be because of public health measures, TB treatment, and possibly vaccination and prevention efforts.

# Conflict and Terrorism Cases

```python
# Global Average Cases Over Time for Conflict and Terrorism
global_avg_conflict_and_terrorism = df.groupby('year')['conflict_and_terrorism'].mean().reset_index()

# Plot Global Average Cases Over Time for Conflict and Terrorism
fig_global = px.line(global_avg_conflict_and_terrorism, x='year', y='conflict_and_terrorism',
                     title='Global Average Conflict and Terrorism Cases Over Time',
                     labels={'conflict_and_terrorism': 'Number of Cases', 'year': 'Year'})

fig_global.update_traces(mode='lines+markers', marker=dict(size=5) , line=dict(color='red') )
fig_global.show()
```

The figure shows the global average number of conflict and terrorism cases from 1990 to 2020. It displays a significant spike in the early 1990s, reaching a peak of over 12,000 cases. This sharp rise represents a sudden and intense surge in conflict and terrorism-related incidents, possibly due to a major global event or series of events. After this peak, the time series shows a rapid decline, followed by a period of relative stability from the mid-1990s to the early 2000s, with only minor fluctuations.

Around 2014, the trend line begins to rise again, indicating an increase in conflict and terrorism cases, which reaches another high point. This change suggests a period of growing global tensions or conflicts. After the mid-2010s, the time series shows a downward trend, reflecting a decrease in the number of conflict and terrorism incidents in recent years.

```python
# Filter Data for the Philippines for Conflict and Terrorism
philippines_conflict_and_terrorism = df[df['country'] == 'Philippines'][['year', 'conflict_and_terrorism']]
# Plot Cases Over Time in the Philippines for Conflict and Terrorism
fig_philippines = px.line(philippines_conflict_and_terrorism, x='year', y='conflict_and_terrorism',
                          title='Conflict and Terrorism Cases in the Philippines Over Time',
                          labels={'conflict_and_terrorism': 'Number of Cases', 'year': 'Year'})

fig_philippines.update_traces(mode='lines+markers', marker=dict(size=5), line=dict(color='red'))
fig_philippines.show()
```

The figure shows the number of conflict and terrorism cases in the Philippines from 1990 to 2020. It highlights a significant spike in the early 1990s, reaching over 2,000 cases, indicating a sudden surge in conflict and terrorism-related incidents. After this peak, there is a rapid decline, followed by a period of fluctuation with smaller peaks in the mid-1990s.

Around the year 2000, the graph shows another sharp increase, suggesting a resurgence of conflict, which is then followed by another decline and period of relative stability. In the mid-2000s to early 2010s, the number of cases fluctuates with moderate increases and decreases, indicating ongoing instability. There is a noticeable rise in cases around 2017, reaching another high point before declining again toward 2020.

# Road Injuries Cases

```python
# Filter the data for the Philippines
df_phil = df[df['country'] == 'Philippines']

# Calculate global average road injuries cases by year
df_global_avg = df.groupby('year')['road_injuries'].mean().reset_index()

# Plot global average road injuries cases over time
fig1 = px.line(
    df_global_avg,
    x='year',
    y='road_injuries',
    title='Global Average Road Injuries Cases Over Time',
    labels={'road_injuries': 'Average Road Injuries Cases', 'year': 'Year'},
    line_shape='spline',
    markers=True
)

# Plot road injuries cases over time (Philippines)
fig2 = px.line(
    df_phil,
    x='year',
    y='road_injuries',
    title='Road Injuries Cases Over Time (Philippines)',
    labels={'road_injuries': 'road injuries Cases', 'year': 'Year'},
    line_shape='spline',
    markers=True
)

fig1.update_layout(template='plotly_dark', title_x=0.5)
fig2.update_layout(template='plotly_dark', title_x=0.5)

# Show plots
fig1.show()
fig2.show()
```

The visualizations compare global and Philippine trends in road injuries over time. The global plot reveals whether road injuries are increasing or decreasing worldwide, offering insights into the effectiveness of global road safety measures. The Philippine-specific plot allows for a comparison with global trends, highlighting any unique challenges or improvements in the country. Any significant changes in these trends can point to the impact of events such as new traffic laws or safety campaigns. These results provide valuable information for policymakers to enhance road safety strategies both globally and locally.


 - Global:

   - From 1990 to around 2008, there is a steady increase in the global average of road injury cases, peaking in 2008.

   - After peaking, the number of cases begins to decrease steadily, continuing this trend until around 2017.

   - Overall: Globally, road injuries increased up until the mid-2000s, after which there was a decline, possibly due to improved road safety measures and regulations. However, there is a slight rise or stabilization in cases after 2017.

 - Philippines:

   - The number of cases starts relatively low in 1990, then slightly decreases until around the 1993-1995.

   - From the late 1990s onwards, there's a steady increase in road injury cases, with a sharp rise after 2007, continuing to climb until 2020.

   - Overall: In the Philippines, unlike the global trend, road injuries have been on the rise, especially after 2005. This may indicate increasing vehicular activity, urbanization, or possibly less effective road safety measures compared to the global average.
